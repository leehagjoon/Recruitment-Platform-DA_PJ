```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from urllib.parse import urlsplit, parse_qs
```

```python
 df = pd.read_csv('./data/application_log2023_final.csv')
```

```python
df.head()
```

```python
def parse_url_components(url):
    if pd.isna(url) or not isinstance(url, str):
        return None, None
    parsed = urlsplit(url.strip())
    return parsed.path, parsed.query

def parse_query_to_dict(query):
    if pd.isna(query) or not isinstance(query, str) or not query.strip():
        return {}
    return {k: v[0] for k, v in parse_qs(query).items()}

df['Path'], df['Query'] = zip(*df['URL'].apply(parse_url_components))
df['Query_Dict'] = df['Query'].apply(parse_query_to_dict)

df.to_csv('processed_url_data.csv', index=False, encoding='utf-8-sig')
```
